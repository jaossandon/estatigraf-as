```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(5, 10))

# Datos de los estratos (profundidad en cm)
estratos = ['Estrato A', 'Estrato B', 'Estrato C', 'Estrato D']
profundidades = [50, 30, 40, 30]
texturas = ['/', '\\', '|', '-']

# Posición inicial de cada estrato en la gráfica
inicio = 0

for i, (estrato, prof, textura) in enumerate(zip(estratos, profundidades, texturas)):
    ax.barh(estrato, prof, left=inicio, color="white", edgecolor="black", hatch=textura)
    inicio += prof

# Ajustes de la gráfica
ax.set_xlabel('Profundidad (cm)')
ax.set_ylabel('Estratos')
ax.set_xlim(0, sum(profundidades))
ax.invert_yaxis()
ax.set_title('Columna Estratigráfica con Texturas')

# Agregar leyenda
legend_patches = [mpatches.Patch(facecolor="white", edgecolor="black", hatch=textura, label=f'{estrato} ({prof} cm)')
                  for estrato, prof, textura in zip(estratos, profundidades, texturas)]
ax.legend(handles=legend_patches, loc='upper right')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
estratos = ['Estrato A', 'Estrato B', 'Estrato C', 'Estrato D']
profundidades = [50, 30, 40, 30]
colores = ['#FFC300', '#FF5733', '#C70039', '#900C3F']

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(6, 8))

# Posición inicial de cada estrato en la gráfica
inicio = 0

for i in range(len(estratos)):
    ax.barh(estratos[i], profundidades[i], left=inicio, color=colores[i], edgecolor="black")
    ax.text(inicio + profundidades[i] / 2, i, f'{estratos[i]}: {profundidades[i]} cm', ha='center', va='center', color='black')
    inicio += profundidades[i]

# Configuración del gráfico
ax.set_xlabel('Profundidad acumulada (cm)')
ax.set_ylabel('Estratos')
ax.set_xlim(0, sum(profundidades))
ax.invert_yaxis()
ax.set_title('Columna Estratigráfica Segmentada')

plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Datos para los estratos
estratos = ['Estrato A', 'Estrato B', 'Estrato C', 'Estrato D']
profundidades = [50, 30, 40, 30]
descripciones = [
    "Grava y arena media - Suelo fértil.",
    "Grava masiva - Estrato bien compactado.",
    "Paquete contenedor de conchal - Depósito sedimentario.",
    "Paquete estéril de grava y arena - Sin materia orgánica."
]
colores = ['#f6c7b6', '#f29e94', '#d94f4f', '#b71b1b']

# Agregar las barras para los estratos
for i in range(len(estratos)):
    fig.add_trace(go.Bar(
        x=[estratos[i]],
        y=[profundidades[i]],
        marker=dict(color=colores[i], line=dict(color='black', width=2), opacity=0.85),
        name=estratos[i],
        text=f'{descripciones[i]} ({profundidades[i]} cm)',
        textposition='outside',
        textfont=dict(size=12, color='black')
    ))

# Añadir líneas guía en el eje Y
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

# Ajustar el layout del gráfico
fig.update_layout(
    title=dict(
        text="Columna Estratigráfica con Mejoras Finales",
        font=dict(size=18, color='darkblue'),
        x=0.5
    ),
    xaxis_title="Estratos",
    yaxis_title="Profundidad (cm)",
    yaxis=dict(autorange="reversed", showgrid=True, gridcolor='LightGrey', zeroline=False),
    xaxis=dict(showgrid=False, zeroline=False),
    showlegend=False,
    bargap=0.1,
    plot_bgcolor='white',
    height=600
)

# Añadir anotaciones para mayor claridad
for i in range(len(estratos)):
    fig.add_annotation(
        x=estratos[i],
        y=profundidades[i] + 5,
        text=descripciones[i],
        showarrow=False,
        font=dict(size=10, color='black'),
        align='center',
        xanchor='center',
        yanchor='bottom'
    )

# Mostrar el gráfico
fig.show()
```
